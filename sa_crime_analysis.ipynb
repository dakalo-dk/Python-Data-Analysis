```python
#All categories of crimes in South Africa
dataset.Category.unique()
```

# Add a sum column at the end of dataset

```python
dataset["Sum"] = dataset.sum(axis=1) #this sums up the row data
dataset.head()
```

```python
dataset.info()
```

# Number of crimes reported per year and over 2005 - 2016 per province

```python
df = (dataset.groupby(by=["Province"],as_index=False).sum()).sort_values(by="Province")
df
```

```python
df.columns
```

```python
df = df.drop(columns=["Province","Sum"])
df
```

```python
Y = list(df.iloc[0])
Y
```

```python
X = df.columns
X
```

# Plot to show reported crimes per province from 2005 to 2016

```python
plt.figure(figsize=(10,5)) #figsize=(colum,row)
plt.plot(X,list(df.iloc[0]),label="EC",linewidth=5)
plt.plot(X,list(df.iloc[1]),label="FS",linewidth=5)
plt.plot(X,list(df.iloc[2]),label="GP",linewidth=5)
plt.plot(X,list(df.iloc[3]),label="KZN",linewidth=5)
plt.plot(X,list(df.iloc[4]),label="LP",linewidth=5)
plt.plot(X,list(df.iloc[5]),label="MP",linewidth=5)
plt.plot(X,list(df.iloc[6]),label="NW",linewidth=5)
plt.plot(X,list(df.iloc[7]),label="NC",linewidth=5)
plt.plot(X,list(df.iloc[8]),label="WC",linewidth=5)

plt.xticks(rotation=90)
plt.ylabel("No. of Crimes")
plt.title("Number of crimes reported per Province from 2005 to 2016")
plt.legend()
plt.show()
```

# Top 10 highest crime Areas in South Africa

```python
crime_areas = (dataset.groupby(by=["Station","Province"],as_index=False).sum()).sort_values(by="Sum",ascending=False)
crime_areas.head(10)
```

# 10 Areas with least crimes in South Africa

```python
crime_areas.tail(10)
```

# Total number of crimes reported per category

```python
crimes = (dataset.groupby(by=["Category"],as_index=False).sum()).sort_values(by="Sum",ascending=False)
crimes
```

# Area where a certain crime was reported the most

```python
top_crime = (dataset.loc[dataset.groupby(by=["Category"],as_index=False).Sum.agg("idxmax")]).sort_values(by="Sum",ascending=False)
top_crime
```

```python
sns.barplot(x="Category", y="Sum",data=top_crime)
plt.xticks(rotation=90)
```

# Area where certain crime was least reported

```python
least_crime = (dataset.loc[dataset.groupby(by=["Category"],as_index=False).Sum.agg("idxmin")]).sort_values(by="Sum",ascending=False)
least_crime
```

```python
sns.barplot(x="Category", y="Sum",data=least_crime)
plt.xticks(rotation=90)
```
